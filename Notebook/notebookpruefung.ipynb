-- Boxplots f√ºr Ausrei√üer-Analyse --

```python
print("\n" + "=" * 60)
print("Boxplots - Ausrei√üer-Analyse")
print("=" * 60)

numeric_cols = df.select_dtypes(include=[np.number])

fig, axes = plt.subplots(2, 4, figsize=(22, 14))
axes = axes.flatten()

for idx, col in enumerate(numeric_cols):
    ax = axes[idx]

    # Boxplot erstellen
    bp = ax.boxplot(df[col], vert=True)

    ax.set_title(f'Boxplot: {col}', fontsize=12, fontweight='bold')
    ax.set_ylabel(col)
    ax.grid(True, alpha=0.3, axis='y')

    # Ausrei√üer besondere Informationen
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = ((df[col] < lower) | (df[col] > upper)).sum() # alle Ausrei√üer au√üerhalb der p-Quantile z√§hlen

plt.tight_layout()
plt.show()


print("\n" + "=" * 60)
print("Detaillierte Ausrei√üer-Analyse")
print("=" * 60)

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers_count = ((df[col] < lower) | (df[col] > upper)).sum()
    outliers_pct = outliers_count / len(df) * 100

    if outliers_count > 0:
        print(f"{col.upper()} ist nicht perfekt normalverteilt!")
        print(f"   Interquartils-Bereich: [{Q1:.0f}; {Q3:.0f}]")
        print(f"   Ausrei√üer-Bereich: (<{lower:.0f}, >{upper:.0f})")
        print(f"   Anteil der Ausrei√üer: {outliers_count} ({outliers_pct:.2f}%)")


print("")
print("=" * 60)
```

### Interpretationen:
- Ausrei√üer = alle Werte oberhalb oder unterhalb der Whister
- der Interquartilsbereich ist der, welcher innerhalb der Box des Box-Plots liegt

### Bivariate Analyse - Korrelationen

-- Korrelationsmatrix --

```python
print("\n" + "=" * 60)
print("Korrelationsmatrix")
print("=" * 60)

# deposit in bin√§r konvertieren, um mit den numerischen Werten arbeiten zu k√∂nnen
df['deposit_binary'] = (df['deposit'] == 'yes').astype(int)

# loan, housing, default ebenso bin√§r konvertieren
df['default_binary'] = (df['default'] == 'yes').astype(int)
df['housing_binary'] = (df['housing'] == 'yes').astype(int)
df['loan_binary'] = (df['loan'] == 'yes').astype(int)

# Korrelationsmatrix berechnen
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
correlation_matrix = df[numeric_cols].corr()

print("\nKorrelationsmatrix:")
print(correlation_matrix.round(3))

print("")
print("=" * 60)
```

-- Heatmap visualisieren --

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            fmt='.2f', square=True, linewidths=1)
plt.title('Korrelationsmatrix der numerischen Variablen', fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()

print("")
print("=" * 60)
```

### Interpretationen:
-> je h√∂her bzw. niedriger der Wert, desto h√∂her auch die Korrelation, also desto St√§rker h√§ngen die Variablen voneinander ab. Zwei Variablen ver√§ndern sich gemeinsam, es besteht ein statistischer Zusammenhang, aber ggf. ohne zwingenden Ursache‚ÄëWirkungs‚ÄëBezug.

- Wert 1:  positiver Zusammenhang der Variablen (beide Attributwerte sind hoch oder niedrig)
- Wert 0: kein linearer Zusammenhang zwischen den Attributwerten
- Wert -1: negativer Zusammenhang der Variablen (ist der eine Attributwert hoch, ist die andere niedrig und umgekehrt)

-> bspw. `duration` und `deposit_binary`: 0.45 -> leicht positiver Zusammenhang

-> bspw. `duration` und `loan_binary`: -0.00 -> kein linearer Zusammenhang

-> bspw. `age` und `housing_binary`: -0.17 -> sehr schacher negativer Zusammenhang



-> in diesem Datensatz sind die Attributwerte meist sehr unabh√§ngig voneinander

-> die gr√∂√üte Abh√§ngigkeit ist zwischen `pdays` und `previous` mit einem Korrelationswert von 0.51

### Behandlung spezieller Werte (pdays, bin√§re Ja/Nein-Felder)

```python
print("="*60)
print("Behandlung spezieller Werte (pdays)")
print("="*60)

# pdays: -1 bedeutet noch nie kontaktiert
no_prev_contact_mask = df_clean['pdays'] == -1
print(f"Anteil 'pdays = -1' (kein fr√ºherer Kontakt): {no_prev_contact_mask.mean()*100:.1f}%")

# Neues bin√§res Feature: fr√ºherer Kontakte? 1-> ja und 0-> nein
df_clean['has_prev_contact'] = (~no_prev_contact_mask).astype(int)

# pdays f√ºr echte Kontakte sinnvoll ver√§ndern (z.B. -1 -> NaN)
df_clean['pdays_clean'] = df_clean['pdays'].replace(-1, np.nan)
```

### Feature Engineering: abgeleitete Variablen

Durch Feature Engineering werden zus√§tzliche Informationen explizit
modelliert, die nicht direkt in einzelnen Variablen enthalten sind.
Beispielsweise kann eine hohe Anzahl von Kontakten auf eine geringere
Abschlusswahrscheinlichkeit hindeuten.

Ebenso fasst das Feature `credit` verschiedene Kreditarten zusammen
und modelliert die allgemeine finanzielle Verpflichtung eines Kunden.


```python
print("="*60)
print("Feature Engineering ‚Äì neue Features")
print("="*60)

# 1) Bin√§res Feature: many_contacts (viele Kontakte in aktueller Kampagne)
df_encoded['many_contacts'] = (df_encoded['campaign'] >= 4).astype(int)
# 2) Interaktionsfeature: contacts_total (Summe aller Kontakte allgemein)
df_encoded['contacts_total'] = df_encoded['campaign'] + df_encoded['previous']

# 3) Verh√§ltnis-Feature: contacts_ratio = previous / (campaign + previous)
df_encoded['contacts_total'] = df_encoded['campaign'] + df_encoded['previous']
df_encoded['contacts_ratio'] = np.where(
    df_encoded['contacts_total'] > 0,
    df_encoded['previous'] / df_encoded['contacts_total'],
    0
)

# 4) Interaktionsfeature: credit (besteht allgemein ein Kredit?)
df_encoded["credit"] = (df_encoded["default_binary"] | df_encoded["loan_binary"] | df_encoded["housing_binary"])
df_encoded["credit"] = df_encoded["credit"].astype(int)

print("Zus√§tzliche Features erzeugt: 'many_contacts', 'contacts_total', 'contacts_ratio', 'credt'")
print("="*60)


df_encoded
```

- contacts_ratio -> Anteil der fr√ºheren Kontakte an der Gesamtzahl der Kontakte (float)

## Teil 3: Unsupervised und Supervised Learning

Ziel ist es, einerseits Kundensegmente ohne Nutzung der Zielvariable
(Unsupervised Learning) zu identifizieren und andererseits vorherzusagen,
ob ein Kunde ein Festgeldkonto abschlie√üt (Supervised Learning).

Ziel des k-Means-Clusterings ist es, Kunden anhand ihrer Merkmalsauspr√§gungen
in homogene Gruppen zu segmentieren, ohne die Zielvariable `deposit` zu verwenden.
Da es sich um numerische, skalierte Daten handelt und eine klare Segmentierung
angestrebt wird, eignet sich k-Means als distanzbasiertes Verfahren besonders gut.

- Da keine Zielvariable vorliegt und zun√§chst unbekannte Strukturen in den Daten identifiziert werden sollen, wird ein Unsupervised-Learning-Ansatz gew√§hlt.
- (Unsupervised Learning eignet sich insbesondere zur explorativen Analyse und Hypothesenbildung.)

### Vorgehen

Zentrale Fragestellung (Supervised Learning):
- K√∂nnen wir vorhersagen, ob ein Kunde einen Term Deposit abschlie√üt (`deposit_bin`), basierend auf soziodemografischen Merkmalen und Kampagneninformationen?

Unsupervised Learning:
- Unterscheidet sich das Verhalten/Profil von Kunden, die abschlie√üen, von denen, die nicht abschlie√üen ‚Äì erkennbar √ºber Cluster in den Feature-Daten?

### Unsupervised Learning (explorative Analyse mit Assoziationsregeln)

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

# Alle bin√§ren Spalten (0 oder 1) im DataFrame ausw√§hlen
binary_cols = [c for c in df_encoded.columns if set(df_encoded[c].unique()).issubset({0, 1})]


# Nur diese Spalten f√ºr Apriori einsetzen
df_binary = df_encoded[binary_cols]
# Alle bin√§ren Spalten in den Datentyp Boolean umwandeln
df_binary_bool = df_binary.astype(bool)


frequent_itemsets = apriori(df_binary_bool, min_support = 0.2, use_colnames = True) # alle Items mit minimalen Support von 20%


rules = association_rules(frequent_itemsets, metric = "confidence", min_threshold = 0.4) # minimale Confidence ist 40%
rules_deposit = rules[rules['consequents'].apply(lambda x: 'deposit_binary' in x)] # zeige nur Beispiele mit consequents = Zielvariable
rules_deposit[["antecedents", "consequents", "support", "confidence", "lift"]].sort_values(
by=["support"], ascending=[False]
)
```

### Interpretation

support: Wahrscheinlichkeit, dass beide Werte gemeinsam in einer Zeile vorkommen

confidence: Wahrscheinlichkeit, dass deposit_binary zutrifft, wenn antecedent auch zutrifft

lift: positiver oder negativer Zusammenhang bzw kein Zusammenhang

`contact_cellular`
-> rund 39 % aller Kunden haben einen Abschluss get√§tigt nachdem sie telefonisch kontaktiert wurden

-> die Wahrscheinlichkeit eines Abschlusses nach telefonischem Kontakt liegt bei ca. 54%, demnach sind mehr als die H√§lfte der Telefonate erfolgreich

-> lift > 1 -> es besteht ein leicht positiver Zusammenhang

`marital_married`

-> nur rund 25 % der verheirateten haben ein Festgeldkonto angelegt

-> die Wahrscheinlichkeit verheiratet zu sein und ein Festgeldkonto zu haben liegt bei 43 %, also weniger als die H√§lfte

-> lift < 1 -> es besteht ein leicht negativer Zusammenhang, bei verheiratete gibt es weniger Abschl√ºsse

### Unsupervised Learning (k-Means Clustering)

- k-Means wird gew√§hlt, da es gut interpretierbare, kompakte Cluster erzeugt und sich f√ºr numerische, skalierte Daten eignet.

### Principal Component Analysis (PCA, Hauptkomponentenanalyse) zur Dimensionsreduktion

```python
from sklearn.decomposition import PCA
from sklearn import decomposition

pca = decomposition.PCA(n_components = 2) # auf 2 Dimensionen reduzieren
pca.fit(X) # Eigenwerte und Eigenvektoren berechnen
df_pca = pca.transform(X)

for i, component in enumerate(pca.components_):
    print(f"{i+1}. Komponente: {100 * pca.explained_variance_ratio_[i]:3.3}% der Varianz")
    print(" + ".join(f"{value:.3} x {name}" for value, name in zip(component, X)))
```

- hier werden hochdimensionale Daten auf zwei Dimensionen reduziert, um sie besser visualisieren und analysieren zu k√∂nnen
- dabei wird die Richtung der h√∂chsten und die der zweith√∂chsten Varianz der Daten analysiert, in linearer Kombination aller Features
- die Varianz sagt aus wie gro√ü die Unterschiede zwischen den Daten sind / wie viel der Gesamtvarianz diese Komponente h√§lt


- daher dass der Varianzanteil relativ klein ist, ist kein dominantes Muster sichtbar
- 1. Varianz: gr√∂√üte Einflussgr√∂√üen sind `contacts-ratio`und `has_prev_contact` mit ca. 0.39
-> scheint durch vergangene Kontakte gepr√§gt zu sein
- 2. Varianz: gr√∂√üte Einflussgr√∂√üen sind `campaign_log1p` und `contacts_total` mit ca 0.49 und 0.44
-> scheint von Kontakten w√§hrend der aktuellen Kampagne gepr√§gt zu sein

```python
print("="*60)
print("Unsupervised Learning ‚Äì K-Means-Clustering")
print("="*60)


df_pca = pd.DataFrame(df_pca, columns=['PC1', 'PC2']) # PCA-Ergebnisse der beiden Hauptkomponenten in DataFrame umwandeln

# die Zielvariable hinzuf√ºgen
df_pca['deposit_binary'] = df_scaled['deposit_binary'].values

# Scatterplot mit F√§rbung nach deposit_binary
sns.scatterplot(
    data=df_pca, 
    x='PC1', 
    y='PC2', 
    hue='deposit_binary',  # farblich nach dieser Variable unterscheiden
    palette='bright', 
)

plt.title('PCA Scatterplot nach deposit_binary')
plt.xlabel('1. Komponente')
plt.ylabel('2. Komponente')
plt.legend(title='deposit_binary')
plt.show()

print("="*60)
```

- durch PCA werden zwei Cluster zusammengefasst

### Cluster-Profiling (Interpretation)

Ziel: pro Cluster die wichtigsten Kennzahlen zeigen (Segmente beschreiben).

Zur Bestimmung der optimalen Clusteranzahl wird die Elbow-Methode, sowie die Silhouett-Analyse verwendet.

```python
print("="*60)
print("Elbow-Methode")
print("="*60)

# Elbow-Methode f√ºr k-Means-Clustering 
inert = [KMeans(n_clusters=k, random_state=42, n_init="auto").fit(df_pca).inertia_ for k in range(1, 10)]
plt.plot(range(1, 10), inert, marker="o")
plt.show()
```

Ab k = 3 flacht die Kurve deutlich ab, sodass
zus√§tzliche Cluster nur noch einen geringen Informationsgewinn liefern.
Daher wird k = 3 als sinnvolle Clusteranzahl gew√§hlt.

- Stelle, wo die Kurve abflacht (Ellbow) = optimale Clusteranzahl
-> hier 3

### Silhouette-Score (kurze G√ºtezahl)
Der Score misst grob: wie gut die Punkte innerhalb eines Clusters zusammenpassen und wie klar die Cluster voneinander getrennt sind (h√∂her = besser).

```python
from sklearn.metrics import silhouette_score

print("="*60)
print("Silhouette-Analyse")
print("="*60)

scores = []
ks = range(2, 9)
for k in ks:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(X)
    scores.append(silhouette_score(X, labels))

plt.plot(list(ks), scores, marker="o")
plt.xlabel("k")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score f√ºr k-Means")
plt.grid(True)
plt.show()
print("\n" + "=" * 60)
```

- Zur Wahl der Clusteranzahl wurde der Silhouette-Score und die Elbow-Methode verwendet. Der h√∂chste Score der Silhouette ergibt sich bei k=2, was auf die beste Clustertrennung in diesem Bereich hinweist. F√ºr h√∂here k sinkt der Score deutlich und bleibt insgesamt niedrig, was eine nur moderate Separierbarkeit der Kundengruppen nahelegt.

- Die Elbow-Methode zeigt jedoch einen deutlichen Knick bei k = 3, was auf eine gute Balance zwischen erkl√§rter Varianz und Modellkomplexit√§t hindeutet.

- F√ºr die weitere Analyse wurde also k=3 gew√§hlt, da dadurch drei interpretierbare Segmente entstehen und die Segmentierung f√ºr die fachliche Auswertung im Vordergrund steht.

soziodemografischen Merkmalen und Kampagnenreaktionen.

### Supervised Learning: Baseline‚ÄëModell (Logistische Regression)

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression


print("="*60)
print("Train-/Test-Split (stratifiziert)")
print("="*60)

# Annahme: X und y wurden in Schritt 2 definiert
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2, # 20% der Daten f√ºr Test, 80% f√ºr Train
    random_state=42,
    stratify=y # Daten haben gleiche relative Klassenh√§ufigkeit wie y
)

print("="*60)
print("Supervised Learning ‚Äì Logistische Regression (Baseline)")
print("="*60)

log_reg = LogisticRegression(max_iter=1000, random_state=42) # iterriert bis maximal 1000 mal
log_reg.fit(X_train, y_train)


print(f"Train-Shape X_train: {X_train.shape}, y_train: {y_train.shape}")
print(f"Test-Shape  X_test : {X_test.shape}, y_test : {y_test.shape}")
print(f"Klassenverteilung gesamt: {y.value_counts(normalize=True).to_dict()}")
print(f"Klassenverteilung Train : {y_train.value_counts(normalize=True).to_dict()}")
print(f"Klassenverteilung Test  : {y_test.value_counts(normalize=True).to_dict()}")
print("="*60)
```

- Die Daten werden in Trainings- und Testdaten aufgeteilt, um die
Generalisierungsf√§higkeit der Modelle zu bewerten.
Ein stratifizierter Split stellt sicher, dass die Klassenverteilung
der Zielvariable `deposit` in beiden Datens√§tzen erhalten bleibt.
-> durch stratify=y haben die Train- und Testdaten die gleiche Verteilung wie die Gesamtdaten (um Verzerrungen zu vermeiden)

- Eine lineare Regression eignet sich nicht f√ºr die Vorhersage der bin√§ren
Zielvariable `deposit`. Stattdessen werden Klassifikationsmodelle (Logistische Regression, Entscheidungsbaum) verwendet, die speziell f√ºr bin√§re Zielvariablen konzipiert sind.

- hier wird statt der Linearen Regression die Logistische Regression angewandt: hier liegt unser Ergebnis zwangsweise zwischen 0 und 1 -> das Ergebnis ist eine Wahrscheinlichkeit

```python
print("Datentypen in X_train:")
print(X_train.dtypes.value_counts())

print("\nNicht-numerische Spalten in X_train:")
non_numeric_train = X_train.select_dtypes(exclude=[np.number]).columns.tolist()
print(non_numeric_train)
```

-> erneute √úberpr√ºfung auf numerischen und nicht-numerische Spalten, da nicht-numerische Spalten zu einem Fehler f√ºhren w√ºrden

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    roc_curve,
    confusion_matrix,
    ConfusionMatrixDisplay
)

print("="*60)
print("Konfusionsmatrix erstellen")
print("="*60)


y_predicted = log_reg.predict(X_test) # Vorhersage mit Test-Daten (0/1)
y_proba = log_reg.predict_proba(X_test)[:, 1] # gesch√§tzte Wahrscheinlichkeit f√ºr 1

acc  = accuracy_score(y_test, y_predicted)
prec = precision_score(y_test, y_predicted)
rec  = recall_score(y_test, y_predicted)
f1   = f1_score(y_test, y_predicted)
auc  = roc_auc_score(y_test, y_proba)

print(f"Accuracy : {acc:.3f}")
print(f"Precision: {prec:.3f}")
print(f"Recall   : {rec:.3f}")
print(f"F1-Score : {f1:.3f}")
print(f"ROC-AUC  : {auc:.3f}")

# Konfusionsmatrix
cm = confusion_matrix(y_test, y_predicted)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['no', 'yes'])
fig, ax = plt.subplots(figsize=(5,5))
disp.plot(cmap='Blues', ax=ax)
ax.set_xlabel("Realit√§t")
ax.set_ylabel("Vorhersage")

plt.title("Konfusionsmatrix ‚Äì Logistische Regression")
plt.show()
```

- Die Logistische Regression dient als **Baseline-Klassifikationsmodell** f√ºr die Vorhersage von `deposit_bin`.
- Wichtige Kennzahlen:
  - **Accuracy** misst den Gesamtanteil korrekter Vorhersagen.
  - **Precision** gibt an, wie viele der als ‚ÄûAbschluss‚Äú vorhergesagten F√§lle tats√§chlich Abschl√ºsse sind.
  - **Recall** misst, wie viele der tats√§chlichen Abschl√ºsse korrekt erkannt werden.
  - **F1-Score** balanciert Precision und Recall.
  - **ROC-AUC** bewertet das Ranking der Wahrscheinlichkeiten unabh√§ngig vom Schwellenwert.
- Die Konfusionsmatrix verdeutlicht das Verh√§ltnis von False Positives und False Negatives.


### Interpretation:
- es wurden 82,8% der F√§lle korrekt vorhergesagt
- das Modell hat eine hohe Precision, die Aussagen stimmen zum Gro√üteil
- dabei werden 20% der Deposits √ºbersehen (siehe Recall)
- Precision und Recall sind gut ausgewogen

### Supervised Learning: Entscheidungsbaum (interpretierbares Modell)

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

print("="*60)
print("3.4 Supervised Learning ‚Äì Entscheidungsbaum")
print("="*60)

tree_clf = DecisionTreeClassifier(
    max_depth=4,
    min_samples_leaf=100,
    random_state=42
)
tree_clf.fit(X_train, y_train)

y_pred_tree = tree_clf.predict(X_test)

acc_tree  = accuracy_score(y_test, y_pred_tree)
prec_tree = precision_score(y_test, y_pred_tree)
rec_tree  = recall_score(y_test, y_pred_tree)
f1_tree   = f1_score(y_test, y_pred_tree)

print(f"Accuracy (Tree) : {acc_tree:.3f}")
print(f"Precision (Tree): {prec_tree:.3f}")
print(f"Recall (Tree)   : {rec_tree:.3f}")
print(f"F1-Score (Tree) : {f1_tree:.3f}")

plt.figure(figsize=(16, 8))
plot_tree(
    tree_clf,
    feature_names=X.columns,
    class_names=['no', 'yes'],
    filled=True,
    rounded=True,
    max_depth=3
)
plt.title("Entscheidungsbaum (max. Tiefe 3 ‚Äì Ausschnitt)")
plt.show()

print("="*60)
```
